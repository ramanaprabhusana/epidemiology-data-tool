# Create Suggestions Document — Epidemiology Evidence Pack

Generates a Word document with structured recommendations for packaging the pipeline output  
into a client-facing "Epidemiology Evidence Pack" that satisfies both the IP student team  
and PharmaACE / InsightACE requirements.

**Output:** `Epidemiology_Evidence_Pack_Suggestions.docx` in the project root  
**Requires:** `python-docx` (`pip install python-docx`)

```python
# Cell 1 — Imports
from pathlib import Path
try:
    from docx import Document
    from docx.shared import Pt, Inches
    print('python-docx OK')
except ImportError:
    print('ERROR: python-docx not installed. Run: pip install python-docx')
```

```python
# Cell 2 — Configuration
OUTPUT_PATH = Path().resolve().parent / "Epidemiology_Evidence_Pack_Suggestions.docx"
print(f'Output: {OUTPUT_PATH}')
```

```python
# Cell 3 — Helpers

def h(doc, text, level=1):
    return doc.add_heading(text, level=level)

def p(doc, text, bold=False):
    para = doc.add_paragraph()
    run = para.add_run(text)
    if bold:
        run.bold = True
    return para

print('Helpers defined')
```

```python
# Cell 4 — Build document

doc = Document()
doc.add_heading("Epidemiology Evidence Pack — Suggestions for Capstone Deliverable", 0)
doc.add_paragraph("PharmaACE IP Project / InsightACE Epidemiology (Beta). Recommendations to satisfy both the student IP team and the client.")
doc.add_paragraph()

h(doc, "1. The Core Issue")
p(doc, "Client side: They want trustworthy epidemiology, confidence, no AI numbers, and something that fits InsightACE. They are not giving you internal data.")
p(doc, "Student side: You need a clear goal, something analytical (not just a list of links), and a concrete deliverable.")
p(doc, "The product should: use only public/source-backed data, add analytics and judgment, and produce client-ready artifacts.")
doc.add_paragraph()

h(doc, '2. The "Great Product": Epidemiology Evidence Pack')
p(doc, "Package the pipeline output into a repeatable product per indication: evidence + confidence + scorecard + white-space + methodology.")
p(doc, "Layer 1 — Extraction: Evidence table, reference links, tool-ready/InsightACE CSV.")
p(doc, "Layer 2 — Analytics: Confidence scoring, source agreement, gaps, optional forecast.")
p(doc, "Layer 3 — Deliverables: One-pager per indication, KPI scorecard, methodology doc, optional dashboard.")
doc.add_paragraph()

h(doc, "3. Concrete Ideas")

h(doc, "3.1 Confidence Score + Rubric (high impact)", level=2)
p(doc, "Rule-based 0–100 score from: source tier, extraction success, completeness, recency, URL verification.")
p(doc, "Deliverable: Confidence on each evidence row + one-page rubric (PDF/Word).")
doc.add_paragraph()

h(doc, "3.2 KPI Scorecard (high impact)", level=2)
p(doc, "For each required metric: coverage (n sources, best value), agreement (do sources agree?), validation readiness.")
p(doc, "Deliverable: KPI scorecard CSV/Excel with metric, best_value, source, range, n_sources, agreement_flag, validation_status.")
doc.add_paragraph()

h(doc, "3.3 White-Space Map (strategic)", level=2)
p(doc, "Coverage matrix: rows = metrics, columns = sources, cell = Has value / Stub / Missing.")
p(doc, "Deliverable: White-space summary table + one page narrative per indication.")
doc.add_paragraph()

h(doc, "3.4 Source Agreement + Reconciliation", level=2)
p(doc, "For metrics with multiple sources: show range, recommended best value, reconciliation rationale.")
p(doc, "Deliverable: Reconciliation table (metric, source_A/B/..., range, recommended, note).")
doc.add_paragraph()

h(doc, "3.5 One-Pager Evidence Summary Per Indication", level=2)
p(doc, "Key numbers, confidence, top sources, main gaps, one-line caveats — per indication.")
p(doc, "Deliverable: One PDF per indication (e.g. CLL_US_Evidence_Summary.pdf).")
doc.add_paragraph()

h(doc, "3.6 Lightweight Forecast with Guardrails", level=2)
p(doc, "Low/base/high scenarios from extracted incidence/prevalence + trend. Always include assumptions note.")
p(doc, "Deliverable: Forecast table + one-page assumptions and limitations.")
doc.add_paragraph()

h(doc, "3.7 Methodology + Reproducibility Pack", level=2)
p(doc, "5–7 page doc: source selection, extraction, confidence scoring, scorecard/white-space production.")
p(doc, "Deliverable: Epidemiology Evidence Pack — Methodology and reproducibility (PDF + config snapshot).")
doc.add_paragraph()

h(doc, "4. How to Package")
p(doc, 'Product name: "Epidemiology Evidence Pack" or "Source-backed epidemiology pack for InsightACE"', bold=True)
p(doc, "For PharmaACE: curated evidence table, confidence-scored dataset, KPI scorecard, white-space summary, optional reconciliation and forecast, methodology pack — all from public sources, no AI-generated numbers, ready to feed InsightACE.")
doc.add_paragraph()

h(doc, "5. Priority Order")
p(doc, "Must-have: Confidence + rubric; KPI scorecard; White-space summary; One-pager per indication; Methodology doc.")
p(doc, "Strong add-on: Source agreement / reconciliation table.")
p(doc, "If time allows: Lightweight forecast; Dashboard heatmap and scorecard view.")
doc.add_paragraph()

doc.add_paragraph("— End of document —")
OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)
doc.save(OUTPUT_PATH)
print(f'Saved: {OUTPUT_PATH}  ({OUTPUT_PATH.stat().st_size:,} bytes)')
```
